```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('data/data.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['cab_type'] = df['cab_type'].replace({'Lyft': 0, 'Uber': 1})
df.head()
```

```python
df = pd.get_dummies(df, columns=['name'])
df.head()
```

```python
# 'Lyft' = 0, 'Uber' = 1
selected_features = ['cab_type', 'distance', 'surge_multiplier', 'source_temp', 'source_clouds', 
                     'source_pressure', 'source_rain', 'source_humidity', 'source_wind', 'name_Black', 
                    'name_Black SUV', 'name_Lux', 'name_Lux Black', 'name_Lux Black XL', 'name_Lyft',
                    'name_Lyft XL', 'name_Shared', 'name_UberPool', 'name_UberX', 'name_UberXL', 'name_WAV']
X = df[selected_features].values
y = df['price'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

model.score(X_test, y_test)
```
